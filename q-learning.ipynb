```
import numpy as np
import random

from random import randrange
```

```
Q = np.zeros((6,6))
R = np.array([[-1,-1,-1,-1, 0,-1],
              [-1,-1,-1, 0,-1,100],
              [-1,-1,-1, 0,-1,-1],
              [-1, 0, 0,-1, 0,-1],
              [ 0,-1,-1, 0,-1,100],
              [-1, 0,-1,-1, 0,100],])

gamma = 0.8
epsilon = 0.2
```

```
def learning():
  random_state = randrange(6) #losowanie stanu początkowego
  while 1 == 1 : #wykonuj dopuki stan nie jest równy 5 (break w 17 linii)
    avalible_actions = np.where(R[random_state] >= 0)[0] #dostępne akcje z bierzącego stanu
    if (random.uniform(0, 1) < epsilon):
      action = random.choice(avalible_actions) #wybór losowej akcji spośród wszystkich możliwych
    else:
      action = np.where(Q[random_state] == np.amax(Q[random_state,avalible_actions]))[0] #Wybór najlepszej akcji
      if (len(action) > 1):
        action = random.choice(list(set(avalible_actions).intersection(action))) #jeżeli jest kikla akcji z taką samą wartością, wylosuj jedną z nich
    if (str(type(action)) == "<class 'numpy.ndarray'>"): #czasami action zamiast być liczbą jest jedno elementową tablicą
      action = action[0]
    avalible_next_actions = np.where(R[action] >= 0)[0] #jeżeli wykonamy akcję i przejdziemy do kolejnego stanu to wybierz możliwe akcje dla nowego stanu
    reward = R[random_state, action] + gamma * np.amax(Q[action,avalible_next_actions]) #funkcja nagrody
    Q[random_state, action] = reward #przypisanie nagrody do Q
    if (action == 5):
      break
    random_state = action #wylosowana akcja staje się bierzącym stanem

def play():
  random_state = randrange(6)
  print(random_state, end = '')
  while random_state != 5:
    action = np.where(Q[random_state] == np.amax(Q[random_state]))[0][0]
    random_state = action
    print('->'+str(action), end = '')    
```

```
Q = np.zeros((6,6))
for i in range(1000):
  learning()

Q = np.round(Q/np.amax(Q)*100,0) #Normalizacja danych na wartości procentowe
Q
```

```
for i in range(20):
  play()
  print('\n')
```
